```python
class Create_node:
    def __init__(self,data):
        self.data = data 
        self.next = None
        self.prev = None
```

```python
class Linked_List:
    
    def __init__(self):
        self.head = None
        
    def print_ll(self):
        if self.head is None:
            print('List is empty')
        else:
            while self.head is not None:
                print(self.head.data,'---->',end = '')
                self.head = self.head.next
                
    def print_ll_rev(self):
        if self.head is None:
            print('List is empty')
        else:
            while self.head.next is not None:
                self.head = self.head.next
            while self.head is not None:
                print(self.head.data)
                self.head = self.head.prev
    def insert_empty(self,data):
        if self.head is None:
            new_node = Create_node(data)
            self.head = new_node
        else:
            print('List is not Empty')
                
    def insert_begin(self,data):
        if self.head is None:
            new_node = Create_node(data)
            self.head = new_node
        else:
            new_node = Create_node(data)
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
            
    def insert_end(self,data):
        if self.head is None:
            new_node = Create_node(data)
            self.head = new_node

        n = self.head
        while n.next is not None:
            n = n.next
        new_node = Create_node(data)
        n.next = new_node
        new_node.prev = n
        
    def add_after(self,data,x):
        if self.head is None:
            print('List is empty')
        else:
            new_node = Create_node(data)
            n = self.head
            while n.next is not None:
                if(n.data==x):
                    break
                n = n.next
            if n is None:
                print('Data is not present')
            else:
                new_node.prev = n
                new_node.next = n.next
                if n.next is not None:
                    n.next.prev = new_node
                n.next = new_node
    def add_before(self,data,x):
        if self.head is None:
            print('Link Link is Empty ')
        elif(self.head == data):
            new_node = Create_node(data)
            self.head = new_node
        else:
            new_node = Create_node(data)
            n = self.head
            while n.next is not None:
                if(n.next.data == x):
                    break
                n = n.next
            if n is None:
                print('Data is not present')
            else:
                new_node.prev = n
                new_node.next = n.next
                n.next.prev = new_node
                n.next = new_node
    def remove_begin(self):
        if self.head is None:
            print('List is empty')
        elif self.head.next is None:
            self.head = None
        else:
            self.head.next.prev = None
            self.head = self.head.next
    
    def remove_end(self):
        if self.head is None:
            print('List is empty')
        elif self.head.next is None:
            self.head = None
        else:
            n = self.head
            while n.next.next is not None:
                n = n.next
            n.next = None
    
    def remove_data(self,data):
        if self.head is None:
            print('List is empty')
        elif(self.head == data):
            self.head.next.prev = None
            self.head = self.head.next
        else:
            n = self.head
            while n.next is not None:
                if(n.data == data):
                    break
                n =n.next
            if n.next is None:
                print('Data is not present')
            else:
                n.prev.next = n.next
                n.prev = None
            
                
            
```
