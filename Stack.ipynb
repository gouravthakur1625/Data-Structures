```python
st=[]
```

```python
def isempty():
    if(st==[]):
        return True
    else:
        return False
```

```python
def pop_element():
    if(st==[]):
        print("Stack is empty")
    else:
        e = st.pop()
        print(e)
```

```python
def push_element():
    el = input("Enter The Element") 
    st.append(el)
    print(st)
```

```python
while True:
    c = int(input("Select the operation: \n1. Push \n2. Pop\n3.Quit\n"))
    if(c==1):
        push_element()
    elif(c==2):
        pop_element()
    elif(c==3):
        break
    else:
        print("Please choose the correct number")
```
