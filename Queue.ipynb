```python
q = []
```

```python
def enque():
    e = input()
    q.append(e)
    print(q)
```

```python
def deque():
    if q==[]:
        print('Queque is empty')
    else:
        e = q.pop(0)
        print(e)      
```

```python
def isnull():
    if(q==[]):
        print('True')
    else:
        print('False')
```

```python
while True:
    c = int(input('Enter the no.\n(1) to Enque\n(2) to deque\n(3) to check queue\n(4) to quit\n'))
    if c==1:
        enque()
    elif c==2:
        deque()
    elif c==3:
        isnull()
    elif c==4:
        break
    else:
        print('Choose the correct no.')
```
