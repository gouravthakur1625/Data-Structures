# Linked List Operation


```python
class Create_Node:
    
    def __init__(self , data):
        self.data = data
        self.ref = None
```

```python
class Linked_list:
    
    def __init__(self):
        self.head = None
    
    def print_ll(self):
        
        if self.head is None:
            print('List is Empty')
        else:
            n = self.head
            while n is not None:
                print(n.data)
                n = n.ref
    
    def insert_begin(self,data):
        new_node = Create_Node(data)
        new_node.ref = self.head
        self.head = new_node
    
    def insert_end(self,data):
        new_node = Create_Node(data)
        if self.head is None:
            new_node = self.head
        else:
            n = self.head
            while n.ref is not None:
                n = n.ref
            n.ref = new_node
        
    def insert_after(self,data,x):
        n = self.head
        while n is not None:
            if x == n.data :
                break
            n = n.ref
        if n is None:
            print('Node is not present in List')
        else:
            new_node = Create_Node(data)
            new_node.ref = n.ref
            n.ref = new_node
            
    def insert_before(self,data,x):
        
        if self.head is None:
            print('List is empty')
        else:
            if x == self.head.data:
                new_node = Create_Node(data)
                new_node.ref = self.head
                self.head = new_node
                
            else:
                n=self.head
                while n is not None:
                    if(n.ref.data ==  x):
                        break
                    n=n.ref
                if n is None:
                    print('Element is not present in List')
                else:
                    new_node = Create_Node(data)
                    new_node.ref = n.ref
                    n.ref = new_node
    def remove_begin(self):
        if self.head is None :
            print('List is empty')
        else:
            self.head = self.head.ref
    
    def remove_end(self):
        if self.head is None:
            print('List is Empty')
        elif self.head.ref is None:
            self.head = None
        else:
            n = self.head
            while n.ref.ref is not None:
                n = n.ref
            n.ref = None
    
    def remove_data(self,data):
        if self.head is None:
            print('List is empty')
            
        elif(self.head.data == data):
            self.head=self.head.ref
            
        else:
            n = self.head
            while n.ref is not None:
                if(n.ref.data == data):
                    break
                n=n.ref
        if n.ref is None:
            print('Data is not present')
        else:
            n.ref = n.ref.ref
                    
            
```

```python
LL1 = Linked_list()
LL1.insert_begin(10)
LL1.insert_begin(23)
LL1.insert_end(55)
LL1.insert_before(145,23)
LL1.remove_end()
LL1.remove_data(26)
LL1.print_ll()
```
